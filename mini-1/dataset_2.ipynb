```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data preprocessing
### Load Dataset 2: wine dataset

```python
# Read the CSV file without dropping the first column
df = pd.read_csv('wine.data', header=None)  # No header, so the first row won't be treated as column names

# Add column names
column_names = ['class', 'Alcohol', 'Malicacid', 'Ash', 'Alcalinity_of_ash', 'Magnesium', 'Total_phenols', 'Flavanoids', 'Nonflavanoid_phenols', 'Proanthocyanins', 'Color_intensity', 'Hue', '0D280_0D315_of_diluted_wines', 'Proline']
df.columns = column_names

# Print the data frame
print(df.head())
print(f"{df.shape[0]} rows\n{df.shape[1]} columns")
```

### check missing value

```python
# Use boolean indexing to find rows containing '?'
rows_with_question_mark = df[df.eq('?').any(axis=1)]

# Check if there are rows with '?'
if not rows_with_question_mark.empty:
    print("DataFrame contains rows with '?'")
    print(rows_with_question_mark)
else:
    print("DataFrame does not contain rows with '?'")
```

# basic statistics on the data

```python
# Drop the 'class' column and calculate the correlation matrix for the feature columns
feature_correlation_matrix = df.drop(columns='class').corr()

feature_correlation_matrix
```

```python
# Create a mapping of feature names to their three-letter abbreviations
abbreviations = {feature: feature[:3] for feature in feature_correlation_matrix.columns}

# Rename the columns and index of the correlation matrix with the three-letter abbreviations
abbreviated_corr_matrix = feature_correlation_matrix.rename(columns=abbreviations, index=abbreviations)

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Draw the heatmap with the three-letter abbreviated feature labels, and adjusted colorbar length
sns.heatmap(abbreviated_corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5,
            xticklabels=True, yticklabels=True, cbar_kws={"shrink": 1})

# Rotate feature labels for better readability and aesthetics
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)

# Title of the heatmap
# plt.title('Adjusted Feature Correlation Heatmap')

plt.tight_layout()
plt.show()
```

```python
# Compute the average absolute correlation for each feature
average_correlation = feature_correlation_matrix.abs().mean(axis=0).sort_values(ascending=False)

# Select the top 5 most important features
top_5_feature_with_feature = average_correlation.head(5)
top_5_feature_with_feature
```


### features

```python
columns_to_process = ['Alcohol', 'Malicacid', 'Ash', 'Alcalinity_of_ash', 'Magnesium', 'Total_phenols', 'Flavanoids', 'Nonflavanoid_phenols', 'Proanthocyanins', 'Color_intensity', 'Hue', '0D280_0D315_of_diluted_wines', 'Proline']

# 2. Create a histogram for each column, dividing into 10 bins based on min and max values
for column in columns_to_process:
    plt.figure(figsize=(8, 4))  # Set the figure size
    
    
    # Create a histogram using plt.hist, dividing into 30 bins based on min and max values
    plt.hist(df[column], bins=30, edgecolor='k')
        
    # Print min, max, mean, and std
    column_stats = df[column].describe()
    print(f'{column} - Min: {column_stats["min"]}, Max: {column_stats["max"]}, Mean: {column_stats["mean"]}, Std: {column_stats["std"]}')
    
    plt.title(f'{column}')
    plt.xlabel('Value Ranges')
    plt.ylabel('Count')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
columns_to_process = ['Total_phenols', 'Flavanoids', 'Hue', 'Malicacid']

# Create subplots with 1 row and 4 columns
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

for i, column in enumerate(columns_to_process):
    # Create a histogram for each column with 30 bins
    axes[i].hist(df[column], bins=30, edgecolor='k')
    
    # Print min, max, mean, and std
    column_stats = df[column].describe()
    print(f'{column} - Min: {column_stats["min"]}, Max: {column_stats["max"]}, Mean: {column_stats["mean"]}, Std: {column_stats["std"]}')
    
    axes[i].set_title(column)
    axes[i].set_xlabel('Value Ranges')
    axes[i].set_ylabel('Count')
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
